```python
import pandas as pd
import os
import shutil
csvfile = 'list_attr_celeba.csv'
images_dir = '/home/irrational/Documents/celeba/'
df = pd.read_csv(csvfile)

img_ids = [imgID for imgID in df["image_id"]]
num_images = len(img_ids)

def GRF(attr_list): # GET RELEVANT FILES
    selected = []
    for img in range(1, num_images+1):
        for attr in attr_list:
            if df[attr][img-1] == -1:
                selected.append(str(img).zfill(6)+".jpg")
                break;
    return selected
def prune(file_list, pwd):
    for img in range(1, num_images+1):
        file = str(img).zfill(6)+".jpg"
        if file not in file_list:
            os.remove(pwd+file)
            



for attr in attr_list:
    if not os.path.isdir(images_dir+attr):
        os.makedirs("Male")
    at = [attr]
    for file in GRF(at):
        shutil.copy(images_dir+"/greyscale/"+file,images_dir+attr+"/"+file)
```

```python
prune(GRF(["Male"]), images_dir+"/greyscale/")
```
