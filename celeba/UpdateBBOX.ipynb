```python
import pandas as pd
from PIL import Image
import numpy as np
```

```python
df = pd.read_csv("list_bbox_celeba.csv")
```

```python
df['center_x'] = df.apply(lambda row: row['x_1'] - (row['width']/2), axis=1)
```

```python
df['center_y'] = df.apply(lambda row: row['y_1'] + (row['height']/2), axis=1)
```

```python
df.head(4)
```

```python
img1 = PIL.Image.open("greyscale/000004.jpg")
```

```python
image_array = np.asarray(img1)
```

```python
import matplotlib.pyplot as plt
```

```python
plt.imshow(image_array)
```

```python
cx, cy = df['center_x'].iloc[0], df['center_y'].iloc[0]
```

```python
figure, axis = plt.subplots()
axis.imshow(image_array)
xs = []
xs.append(cx)
ys = []
ys.append(cy)
axis.scatter(xs, ys, c=xs)
```
