```python
import numpy as np
import pandas as pd
import PIL
import matplotlib.patches as patches
import matplotlib.pyplot as plt
import tensorflow as tf
from sklearn.model_selection import train_test_split
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense,Dropout,Flatten,Conv2D,MaxPooling2D
from tensorflow.keras.preprocessing import image
```

```python
attr_data_path = "list_attr_celeba.csv"
images_data_path = "C:/Users/foreh/Documents/celeba/greyscale"
images_data_size = 20000 

width = 178
height = 218
x_ = 100                            # new value of x
image_size_ratio = width / height    # dimensions ratio
y_ = int(image_size_ratio * x_)  

original_image_size = (width,height)
```

```python
df_org = pd.read_csv(attr_data_path)
df = df_org[:images_data_size]
```

```python
images_data = list()
for idx in range(df.shape[0]):
    path = "{}/{}".format(str(images_data_path),str(df.iloc[idx].image_id))
    image = PIL.Image.open(path)
    image_array = np.asarray(image)
    images_data.append(image_array)
images_data = np.array(images_data)
```

```python
plt.imshow(images_data[3000])
def image_from_index(index, size=(original_image_size), path=images_data_path):
    path = "{}/{}".format(str(path),str(df.iloc[index].image_id))
    # to read the image
    image = PIL.Image.open(path).resize(size)
    image_array = np.asarray(image)
    
    return image_array
```

```python
# fuction to get a list of all key points of the face
def image_attr_list(index, df=df):
    # box dictionary
    attr_list = [df.iloc[index].Straight_Hair,
                   df.iloc[index].Wavy_Hair,
                   df.iloc[index].Bangs,
                   df.iloc[index].Wearing_Hat]
    
    return attr_list
```

```python
print(image_attr_list(5))
plt.imshow(images_data[5])
```

```python
images = images_data # list of array of images
labels = df[["image_id", "Bangs", "Straight_Hair", "Wavy_Hair", "Wearing_Hat"]]  # dataframe of image features
print(labels)
```

```python
train_images,test_images,train_labels,test_labels = train_test_split(images,labels,test_size=0.3,random_state=45)
print(train_labels)
```

```python
y_test=test_labels.drop(["image_id"],axis=1)
y_train=train_labels.drop(["image_id"],axis=1)

X_test=test_images
X_train=train_images
#X_train = tf.expand_dims(X_train, axis=-1)
print("Samples Training Set:",len(X_train))
```

```python
X_test, X_val, y_test, y_val=train_test_split(X_test, y_test,test_size=0.5,random_state=42)
print("Samples in Validation Data:",len(X_val))
print("Samples in Test Data:",len(X_test))
print(X_val[1])
plt.imshow(X_val[1])
```

```python
x_ = original_image_size[0]
y_ = original_image_size[1]
print(y_,x_,1)
```

```python
model = Sequential()
model.add(tf.keras.Input(shape=(y_, x_,1)))
model.add(Conv2D(filters=8,kernel_size=(4,2),padding = 'same', activation = 'tanh', kernel_regularizer='l1'))
## changed from (5,5)->(4,2)

model.add(Conv2D(filters=8, kernel_size=(2, 3), padding='valid', activation='tanh', kernel_regularizer='l1'))
model.add(MaxPooling2D(pool_size=2))
model.add(Dropout(0.4))
model.add(tf.keras.layers.LocallyConnected2D(filters=3,kernel_size=(1,2),padding = 'valid',activation = 'tanh', kernel_regularizer='l1'))
## changed from (3,3)->(1,3)

model.add(Conv2D(filters=16, kernel_size=(3, 3), padding='same', activation='tanh'))
model.add(MaxPooling2D(pool_size=8))
model.add(Flatten())
model.add(Dense(16, activation='tanh', kernel_regularizer='l1'))
model.add(Dropout(0.3))
#
model.add(Dense(4, activation='tanh', kernel_regularizer='l1'))
```

```python
model.summary()
```

Empty box

```python
model.compile(optimizer = "adam",loss="hinge",metrics = ['accuracy'])
```

```python
checkpoint_filepath = 'checkpoint'
model_checkpoint_callback = tf.keras.callbacks.ModelCheckpoint(
    filepath=checkpoint_filepath,
    save_weights_only=True,
    monitor='val_accuracy',
    mode='max',
    save_best_only=False)
```

```python
training_process = model.fit(X_train,y_train,epochs = 20,validation_data = (X_val,y_val),batch_size=2,shuffle = True, callbacks=[model_checkpoint_callback])
```

```python
model.evaluate(X_test,y_test)
model.save_weights('./saved_models/model200')
```

```python
model2 = model
model2.load_weights('saved_models/model1')
```

```python
training_process = model2.fit(X_train,y_train,epochs = 11,validation_data = (X_val,y_val),batch_size=8,shuffle = True)
```

```python
model2.evaluate(X_val, y_val)
```
